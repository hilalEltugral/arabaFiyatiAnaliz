```python
y=df["price"].values #numpy dizilerine çevirdik
x=df.drop("price",axis=1).values #price hariç tüm özellikleri al
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape #5 özelliğimiz var.
```

```python
# Model oluşturuyoruz. 4 katamn ekledik. 1 Çıkışlı
model = Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse") # daha iyi performans için adam kullandık
```

```python
#modelimizi eğitiyoruz..
model.fit(x=x_train, y = y_train,validation_data=(x_test,y_test),batch_size=250,epochs=300)
```

```python
kayipVerisi = pd.DataFrame(model.history.history)
kayipVerisi.head()
```

```python
kayipVerisi.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahminDizisi = model.predict(x_test)
tahminDizisi
```

```python
mean_absolute_error(y_test,tahminDizisi)
```
